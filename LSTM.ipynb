```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Define data and target
data = np.array([[i for i in range(100)]], dtype=float)
target = np.array([[i for i in range(1, 101)]], dtype=float)

# Reshape data for LSTM
data = data.reshape((1, 1, 100))
target = target.reshape((1, 1, 100))

# Test data
x_test = np.array([i for i in range(100, 200)]).reshape((1, 1, 100))
y_test = np.array([i for i in range(101, 201)]).reshape((1, 1, 100))

# Build the LSTM model
model = Sequential()
model.add(LSTM(100, input_shape=(1, 100), return_sequences=True))
model.add(Dense(100))

# Compile the model
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(data, target, epochs=1000)

# Make predictions
predictions = model.predict(data)
print(predictions)
```
