```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean1 = 10
sd1 = 1
lower_bound = -20
upper_bound = 20
x = np.arange(lower_bound, upper_bound, 0.001)
y = norm.pdf(x, 10, 1)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, y)
ax.set_title('Bell curve')
plt.style.use('fivethirtyeight')
plt.show()
```
